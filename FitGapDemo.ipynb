```python
import pandas as pd
import numpy as np
import sqlalchemy as sq
```

```python
myeng = sq.create_engine('mssql://DESKTOP-DKEG4M9\\SQLEXPRESS/pyTest?trusted_connection=yes&driver=ODBC+Driver+13+for+SQL+Server')
```

```python
source_Schema = "dbo"
source_Table = "pyData"

df_pyData = pd.read_sql_table(source_Table, myeng, source_Schema)
df_pyData
```

```python
df_Metric = pd.DataFrame()

df_Metric['minLen'] = df_pyData.fillna(np.NAN).astype(str).apply(lambda x:x.str.len()).min()
df_Metric['maxLen'] = df_pyData.fillna(np.NAN).astype(str).apply(lambda x:x.str.len()).max()
df_Metric['totNumeric'] = df_pyData.fillna(np.NAN).astype(str).apply(lambda x:x.str.isnumeric()).sum()
df_Metric['totNull'] = df_pyData.apply(lambda x:x.isnull()).sum()
df_Metric['totBlank'] = df_pyData.astype(str).apply(lambda x:x.str.strip() == "").sum()
df_Metric['percNullOrBlank'] = ((df_Metric['totNull'] + df_Metric['totBlank']) / df_pyData.shape[0]) * 100.0
```
